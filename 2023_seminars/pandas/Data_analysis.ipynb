# Network analysis

```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from( ['Luca','Andrea','Sara','Carlo','Veronica'] )
```

```python
G.nodes()
```

```python
G.add_node('Giovanni')

print("The nodes of G are: ", G.nodes())
```

```python
G.add_edges_from(
    [ ('Luca','Sara'), ('Andrea','Luca'), ('Carlo','Veronica'), ('Sara','Veronica'), ('Giovanni','Andrea') ]
)
```

```python
G.edges()
```

```python
G.add_edge('Veronica','Lucia')
```

```python
G.nodes()
```

```python
G.nodes['Lucia']
```

```python
for n in G.nodes():
    
    if n[0]=='L':
        G.nodes[n]['age']=24
    else:
        G.nodes[n]['age']=28
```

```python
#show the nodes with their age
print(G.nodes(data=True))
```

```python
G.remove_edge('Giovanni', 'Andrea')
```

```python
G.nodes()
```

```python
G.remove_node('Giovanni')
```

```python
G.nodes()
```

```python
G.degree()
```

# Analisi rete Deezer

```python
Deezer_net = nx.Graph()
```

```python
import csv

with open('./deezer_clean_data/HR_edges.csv', encoding='UTF-8') as f:
    
    lettore = csv.reader(f, delimiter=',')
    
    for riga in lettore:
        
        a = riga.strip().split(',')
        node1 = int(a[0])
        node2 = int(a[1])
        Deezer_net.add_edge(node1, node2)
    
    
    
```

```python
fh = open('./deezer_clean_data/HR_edges.csv', encoding='UTF-8')

fh.readline()

for line in fh.readlines():
    
    a = line.strip().split(',')
    node1 = int(a[0])
    node2 = int(a[1])
    Deezer_net.add_edge(node1, node2)

fh.close()
```
