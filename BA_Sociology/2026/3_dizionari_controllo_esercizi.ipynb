```python
elezioni = {

    'Elena Ferrante': 301,
```

1. Modifica il dizionario e aggiungi una nuova chiave `Bianche/Nulle` con un valore di `48`. Stampa il dizionario.


2. Trova il valore massimo dei voti ottenuti da un singolo candidato e stampalo con un `print`.


3. Calola la percentuale dei voti ottenuti da Elena Ferrante, sul totale, e stampa con un `print` il nome della candidata seguito dalla percentuale. 


4. Scrivi un programma in Python che controlla quanti candidati ci sono che hanno nome uguale a 'Paolo' e stampa il risultato finale con un `print`.


5. Scrivi un programma in Python che stampa consecutivamente i nomi di ogni candidato e la percentuali di voti che ha ottenuto sul totale.



````python
elezioni = {
    'Laura Morante': 232,
    'Carlo Mazzacurati': 198,
    'Paolo Giordano': 202,
    'Paolo Cognetti': 157,
    'Elena Ferrante': 301,
}

#scrivi qui la soluzione 

#aggiungo una chiave al dizionario
elezioni['Bianche/Nulle']=48
print(elezioni)
print()

#trovo il valore massimo dei voti
massimo_voti = max(elezioni.values())
print('Il valore massimo dei voti è:', massimo_voti)
print()

#calcolo la percentuale di voti di Elena Ferrante
totale_voti = sum(elezioni.values())
perc_ferrante = 100*elezioni['Elena Ferrante']/totale_voti
print('Elena Ferrante:', perc_ferrante)
print()

#conto quanti candidati si chiamano Paolo
n_paolo=0
for nome in elezioni:
    if nome.startswith('Paolo'):
        n_paolo+=1
        
print('Ci sono',n_paolo,'candidati che si chiamano Paolo')
print()

#stampo tutti i nomi con la percentuale di voti ottenuti
for nome in elezioni:
    perc_voti = 100*elezioni[nome]/totale_voti
    print(nome+', voti:',round(perc_voti,2),'%')
````

# Esercizio 4

Vogliamo simulare il risparmio mensile di una persona che mette da parte i propri risparmi in un conto corrente.

Hai a disposizione:

```python
saldo_iniziale = 0.0
versamento_mensile = 150.0
interesse_annuo = 0.03  
obiettivo = 5000.0
```

1. Calcola il **tasso di interesse mensile** `interesse_mensile` a partire dal tasso annuo (usa la divisione per 12) e inizializza le variabili necessarie (`saldo`, `mesi`).

2. Usa un **ciclo `while`** per simulare mese per mese: ad ogni iterazione aggiungi `versamento_mensile` al saldo e poi applica l’interesse mensile al nuovo saldo.

3. Fai terminare il ciclo quando il saldo ha raggiunto o superato `obiettivo`.

4. Al termine, stampa il numero totale di mesi necessari e il saldo finale.

````python
saldo_iniziale = 0.0
versamento_mensile = 150.0
interesse_annuo = 0.03   # 3%
obiettivo = 5000.0

#scrivi qui la soluzione
interesse_mensile = interesse_annuo/12

saldo, mesi = saldo_iniziale, 0

while saldo<obiettivo:
    
    mesi+=1
    
    saldo += versamento_mensile
    saldo += saldo*interesse_mensile
    
    print(mesi, saldo)

print()
print('Dopo',mesi,'mesi')    
print('Il saldo finale è', saldo)
````
