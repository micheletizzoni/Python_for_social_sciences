# Esercizio 2

Secondo le leggi della fisica classica: 
- un oggetto in movimento ha un'energia cinetica definita dalla formula: $E = \frac{1}{2} mv^2$ dove $m$ e $v$ sono la massa e la velocità dell'oggetto. 
- un oggetto posto ad altezza $h$ rispetto al suolo ha un'energia potenziale definita dalla formula: $U = g h m$ dove $m$ è la massa e $g$ è l'accelerazione di gravità.

1. Data la massa `m` e la velocità `v`, scrivi in Python una formula che calcola l'energia cinetica `E` di un oggetto in movimento. Stampa il risultato con un `print`. 


2. Dati i valori di `m`, `h` e `g`, scrivi in Python una formula che calcola l'energia potenziale `U` di un oggetto. Stampa il risultato con un `print`. 


3. Confronta l'energia cinetica e quella potenziale trovate, e scrivi un'espressione booleana che restituisce True solo se l'energia cinetica è **minore** di quella potenziale. Stampa il risultato con un `print`. 


```python
m = 5 #massa in kg
v = 5.6 #velocità m/s
h = 2.5 #altezza dal suolo m
g = 9.81 #accelerazione di gravità m/s^2

# calcolo l'energia cinetica
E = (1/2)*(m*v**2)
print("L'energia cinetica è", round(E,4))
print()
# calcolo l'energia potenziale
U = g*h*m
print("L'energia potenziale è", round(U,2))
print()
cond = E<U
print("L'energia cinetica è minore di quella potenziale?", cond)
```

# Esercizio 3

In demografia l'indice di vecchiaia (old age index) misura il rapporto tra la popolazione over 65 e quella under 5 di un paese.

Indichiamo con `pop_ita` la popolazione italiana nel 2023 e con `pop_5` il numero di bambini di età inferiore ai 5 anni, e con `pop_65` il numero di persone di età superiore ai 65 anni.

1. Calcolare la percentuale di popolazione corrispondente a `pop_5` e `pop_65`;
2. Calcolare l'indice di vecchiaia dell'Italia.

Stampa i risultati con un `print`.

```python
pop_ita = 58997201
pop_5 = 2085244
pop_65 = 14181307

# calcolo la percentuale di popolazione

perc_5 = pop_5/pop_ita*100
print("La percentuale di popolazione italiana di età inferiore a 5 anni è "+str(round(perc_5,2))+"%")

print()

perc_65 = pop_65/pop_ita*100
print("La percentuale di popolazione italiana di età superiore a 65 anni è "+str(round(perc_65,2))+"%")

print()

#calcolo l'indice di vecchiaia
old_age_index = pop_65/pop_5
print("L'indice di vecchiaia dell'Italia è",round(old_age_index,2))
```

# Esercizio 4

In economia del lavoro, il salario minimo orario e il numero di ore lavorate determinano il reddito lordo mensile di una persona. Supponiamo che uno stato voglia verificare se una persona rientra o meno in una fascia di **basso reddito** rispetto a una certa soglia.

1. Definisci una variabile intera `ore_lavorate` che rappresenta il numero di ore lavorate in un mese (ad esempio 160).
2. Definisci una variabile `salario_orario` di tipo `float` che rappresenta il salario minimo orario (ad esempio 9.5 euro).
3. Calcola il reddito lordo mensile `reddito_lordo` come prodotto tra `ore_lavorate` e `salario_orario`.
4. Definisci una variabile `soglia_basso_reddito` di tipo `float` (ad esempio 1200.0) che rappresenta la soglia di basso reddito.
5. Crea una variabile booleana `is_basso_reddito` che valga `True` se `reddito_lordo` è **minore** della soglia di basso reddito, e `False` altrimenti.
6. Stampa tutte le variabili (`ore_lavorate`, `salario_orario`, `reddito_lordo`, `soglia_basso_reddito`, `is_basso_reddito`) usando `print` per verificare i risultati.

```python
# scrivi qui la tua soluzione per l'Esercizio 4

# esempio di valori di partenza (puoi modificarli se vuoi)
ore_lavorate = 160      # int
salario_orario = 9.5    # float
soglia_basso_reddito = 1200.0  # float

# calcolo reddito_lordo 
reddito_lordo = ore_lavorate*salario_orario

#calcolo condizione basso reddito
is_basso_reddito = reddito_lordo<soglia_basso_reddito

print("Ore lavorate:\t", ore_lavorate)
print("Salario orario:\t", salario_orario)
print("Soglia reddito:\t", soglia_basso_reddito)
print("Reddito lordo:\t", reddito_lordo)
print("Il reddito è inferiore alla soglia di basso reddito?", is_basso_reddito)
```
